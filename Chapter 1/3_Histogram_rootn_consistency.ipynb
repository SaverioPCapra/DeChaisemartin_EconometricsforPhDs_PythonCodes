```python
import numpy as np 
import matplotlib.pyplot as plt
```

The following code basically simulates the Central Limit Theorem

```python
# I initialize an empty array
A = []

# I start a for loop with 200 iterations
for i in range(200):
    # I generate 1000 observations from the uniform distribution
    u = np.random.uniform(size = 1000)
    # I sum the observations
    mean_u = u.mean()
    # I calculate the scaled mean
    n = np.sqrt(1000)*(mean_u-0.5)
    # I append the scaled mean to the list
    A.append(n)
```

```python
# I plot the array as an histogram
plt.hist(A)
plt.show()
```

As expected, the histogram looks like a bell.
